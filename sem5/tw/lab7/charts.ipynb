```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# get averages
def get_avg(df):
    return df.groupby(['N'], as_index=False).mean()

df_asym = pd.read_csv("asym.csv")
df_waiter_forks = pd.read_csv("waiter_forks.csv")
df_waiter_waiter = pd.read_csv("waiter_waiter.csv")

avgs_asym = get_avg(df_asym)
avgs_waiter_forks = get_avg(df_waiter_forks)
avgs_waiter_waiter = get_avg(df_waiter_waiter)
```

### Wykres dla pojedy≈Ñczych akcji

```python
plt.figure(figsize = (12,8))

plt.plot(avgs_asym['N'], avgs_asym['time'], label="Asym - fork")
plt.plot(avgs_waiter_forks['N'], avgs_waiter_forks['time'], label="Waiter - fork")
plt.plot(avgs_waiter_waiter['N'], avgs_waiter_waiter['time'], label="Waiter - waiter")

plt.title("Average wait time for a single action for both methods")
plt.xlabel("N - amount of philosophers")
plt.ylabel("avg(time) [ms]")

plt.yscale("log")
plt.grid()
plt.legend()
plt.show()
```

### Wykres dla ca≈Çego cyklu jedzenia - 2x widelec dla asym, 2x widelec + kelner dla waiter

```python
plt.figure(figsize = (12,8))

plt.plot(avgs_asym['N'], 2*avgs_asym['time'], label="Asym - 2*fork")
plt.plot(avgs_waiter_forks['N'], 2*avgs_waiter_forks['time'] + avgs_waiter_waiter['time'], label="Waiter - 2*fork + waiter")

plt.title("Average wait time for a whole eating cycle")
plt.xlabel("N - amount of philosophers")
plt.ylabel("avg(time) [ms]")

plt.yscale("log")
plt.grid()
plt.legend()
plt.show()
```
